## Joinplot y Pairplot
Joinplot, permite crear graficos con varios diagramas
Pairplot, genera graficas de data especificada, muy utilzada para realizar analisis iniciales de la informacion

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
# Se puede añadir una segmentacion
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
# Se puede especificar el tipo de diagrama
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde')
plt.show()
```

```python
# Se puede añadir barras para las barras de los ejes
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True)
plt.show()
```

```python
# asi mismo se puede cambiar el estilo de las graficas
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True, marginal_kws=dict(bins = 25, fill = False, multiple='dodge' ))
plt.show()
```

### pairplot
pairplot facilita un resumen de graficas que analiza las variables numericas del insumo

```python
sns.pairplot(data=tips)
plt.show()
```

```python
# Segementar
sns.pairplot(data=tips, hue='sex')
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm')
plt.show()
```

```python
# Para no mostrar los graficos duplicados
sns.pairplot(data=tips, hue='sex',corner=True)
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex',corner=True, kind='scatter')
plt.show()
```
