## Set
Comando que permite cambiar el estilo de las graficas

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x = ['A','B','C'], y = [1,3,2])
# Al ser creado en base a matplotlib afecta directamente a Seaborn
plt.show()
```

```python
# Se le asignan estilos a las graficas
sns.set(style='dark')
sns.barplot(x = ['A','B','C'], y = [1,3,2])
plt.show()
```

```python
# Se puede cambiar el color de las graficas
sns.set(style='whitegrid', palette='pastel')
sns.barplot(x = ['A','B','C'], y = [1,3,2])
plt.show()
```

```python
# Se puede cambiar la fuente
sns.set(style='dark', palette='dark', font='Verdana')
sns.barplot(x = ['A','B','C'], y = [1,3,2])
plt.show()
```

```python
# Y el tama√±o de la fuente
sns.set(style='dark', palette='dark', font='Verdana', font_scale=2)
sns.barplot(x = ['A','B','C'], y = [1,3,2])
plt.show()
```
