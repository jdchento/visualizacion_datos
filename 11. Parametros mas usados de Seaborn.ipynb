## Funciones mas utilizadas en Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Seaborn tiene dataset que permite utilizar para visualizar sus graficas
tips = sns.load_dataset('tips')
print(tips)
```

```python
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
# Identifica cual grafica se acomoda mas a cada informacion
sns.displot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
# hue permite incluir un agrupamiento
sns.displot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex')
plt.show()
```

```python
# kind define con cual grafico se puede trabajar
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde')
plt.show()
```

```python
# Legend permite ocultar o mostrar la legenda
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=False)
plt.show()
```

```python
# Palette cambiar la paleta de colores
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=False, palette='pastel')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=False, palette='pastel', alpha=0.25)
plt.show()
```
