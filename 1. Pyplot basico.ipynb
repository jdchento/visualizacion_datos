## MATPLOTLIB

Matplotlyb esta basada en Numpy, por lo que es muy rapida.

Su sintaxis se asemeja a la de Mathlab


```python
import matplotlib.pyplot as plt
import numpy as np
```

linspace es el metodo para crear una array

```python
x = np.linspace(0, 5, 11)
y = x**2
```

### PLOT
para crear una grafica

```python
plt.plot(x,y)
plt.show()
```

Se pueden hacer modificaciones en la grafica cambiandole los parametros como se presenta a continuacion.

La forma de asignacion es muy similar a Mathlab, siendo la primera letra haciendo referencia al color

```python
plt.plot(x, y, 'rx')
plt.show()
```

```python
plt.plot(x, y, 'bo')
plt.show()
```

```python
plt.plot(x, y, 'gs-')
plt.show()
```

```python
plt.plot(x, y, 'D:')
plt.show()
```

### HIST

```python
plt.hist(x)
plt.show()
```

### PIE

```python
plt.pie(x)
plt.show()
```

### SCATTER

```python
plt.scatter(x, y)
plt.show()
```

### BOXPLOT

```python
plt.boxplot(x)
plt.show()
```
