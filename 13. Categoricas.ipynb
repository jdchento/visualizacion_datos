## Graficos para visualizar datos con variables categoricas

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

```python
# Se puede cambiar el eje
sns.countplot(data=tips, y='day', hue='sex')
plt.show()
```

### Categoricas con grafico de densidad

```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex')
plt.show()
```

```python
# Para analizar mejor los grupos
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
# De igual manera se pueden utilizar las propiedades de matplotlib
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

### Categoricas con grafico de densidad con mejor visualizacion

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

### Distribucion de los datos y concentracion

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

### Se pueden combinar

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, marker='<')
#Se le asigna otro color
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, color='0')
plt.show()
```

### Grafico de violin

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, palette='pastel')
plt.show()
```

```python
# Con split se pueden combinar
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True, palette='pastel')
plt.show()
```

### Con catplot se puede graficar cualquier grafica categorica

```python
plt.figure(figsize=(6,6))
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, kind='violin')
plt.show()
```

```python
# Con col se puede añadir otra variable, como la hora de la comida
plt.figure(figsize=(6,6))
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, col='time')
plt.show()
```
