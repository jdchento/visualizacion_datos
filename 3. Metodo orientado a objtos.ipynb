## Orientado a objetos

Permite una mejor interaccion con las graficas ya que por Pyplot es limitado


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
fig = plt.figure() # Se crea una variable que tendra la figura
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9]) # Posicion que va a tener el lienzo y extension del lienzo
axes.plot(x, y, 'b')
fig.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes2 = fig.add_axes([0.2, 0.6, 0.4, 0.3])

axes.plot(x, y, 'b')
axes2.plot(y, x, 'r')
axes2.set_facecolor('gray') # Se le asigna el color al fondo de la segunda grafica
fig.show()
```

Una figura contiene una grafica (Axes) y este unos ejes (Axis) que los definen
