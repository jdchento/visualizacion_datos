## Subplots

Tiene las bondades del metodo orientado a objetos, que permite modificar de mejor manera las graficas.
Subplot sirve para incluir en una misma salida varias graficas.


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x, y, 'b')
```

```python
#Para definir el numero de graficas
fig, axes = plt.subplots(nrows=1, ncols=2)
# El axes anterior devuelve un Array, por lo que es necesarrio accerder por slide
axes[0].plot(x, y, 'b')
axes[1].plot(y, x, 'r')
```

```python
#Tambien se puede acceder de la siguiente manera, por medio de una tupla
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, y, 'b')
ax2.plot(y, x, 'r')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4)# Tambien se puede escribir solo 2, 4 . Entendiendo que el primero hace referencia a filas y el segundo a columnas
axes[0, 0].plot(x, np.cos(x))
axes[0, 1].plot(x, np.sin(x))
axes[0, 2].plot(x, np.tan(x))
axes[0, 3].plot(x, np.arccos(x))
```

```python

fig, axes = plt.subplots(nrows=2, ncols=4)
axes[0, 0].plot(x, np.cos(x))
axes[0, 1].plot(x, np.sin(x))
axes[0, 2].plot(x, np.tan(x))
axes[0, 3].plot(x, x)
axes[1, 0].plot(x, np.arccos(x))
axes[1, 1].plot(x, np.arcsin(x))
axes[1, 2].plot(x, np.arctan(x))
axes[1, 3].plot(x, x)

# Para visualizar de una manera mas ordenada las graficas
fig.tight_layout()
```
