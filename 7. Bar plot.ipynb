## Bar plot
Es la funcion que permite hacer diagrama de barras que son muy utilizadas para graficar variables categoricas


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000, 800, 900, 1000, 300]
```

```python
plt.bar(country, population)
plt.show()
```

```python
# Modificacion del ancho 
plt.bar(country, population, width=0.5)
plt.show()
```

```python
# Modificacion de colores
plt.bar(country, population, width=0.5, color=['red', 'orange'])
plt.show()
```

```python
# Modificacion de las etiquetas
plt.bar(country, population, width=0.5, color=['red', 'orange'])
# xticks permite entrar a la configuracion del eje x
# se utiliza el arange para sacar el indice de cada barra
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'), rotation=45)
plt.show()
```

```python
# Se generan barras horizontales
plt.barh(country, population, color=['red', 'orange'])
plt.show()
```
