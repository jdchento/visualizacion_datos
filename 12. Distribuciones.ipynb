## Graficos para visualizar datos con distribuciones

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
# Distribucion de los datos
# Las personas principalmente dan 2 dolares de propina
sns.histplot(data=tips, x='tip')
plt.show()
```

```python
# bins sirve para definir el numero de columnas
sns.histplot(data=tips, x='tip', bins=5)
plt.show()
```

```python
# cumulative tiene la funcion de hacer acomulativo
sns.histplot(data=tips, x='tip', bins=5, cumulative=True)
plt.show()
```

```python
# hue para clasificar los valores de la graifca
sns.histplot(data=tips, x='tip', bins=5, cumulative=False, hue='sex')
plt.show()
```

```python
# Estadisticas
# frequency
# probability
# percent
# density
# count
sns.histplot(data=tips, x='tip', bins=5, cumulative=False, hue='sex', stat='count')
plt.show()
```

```python
# Para separar los grupos conformados anteriomente
# stack
# dodge (uno al lado al otro)
# layer (por defecto)
# fill
sns.histplot(data=tips, x='tip', bins=5, cumulative=False, hue='sex', stat='count', multiple='dodge')
plt.show()
```

### Grafico de densidad

```python
sns.kdeplot(data=tips, x='tip', hue='sex')
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=True)
plt.show()
```

```python
# tambien sirve shade, para rellenar
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, fill=True)
plt.show()
```

```python
# bd_ajust sirve para ajustar los ejes
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, fill=True, bw_adjust=1)
plt.show()
```

### Grafico escalonado

```python

sns.ecdfplot(data=tips, x='tip', hue='sex')
plt.show()
```

```python
# se puede cambiar el motodo estadico como se mostro anteriormente
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
plt.show()
```

### Displot es la funcion que por defecto tiene los graficos que pueden ser utilizados para visualizar la distribucion de los datos

```python
sns.displot(data=tips, x='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='kde')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='ecdf')
plt.show()
```

```python
# Igualmente recibe los parametros de las anteriores funciones
sns.displot(data=tips, x='tip', hue='sex', kind='hist', multiple='stack')
plt.show()
```
