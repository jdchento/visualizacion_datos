## Colores y estilos


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
# Muestra todos los estilos disponibles
print(plt.style.available)
```

```python
# Cuadrilla
plt.style.use('bmh')
fig, axes = plt.subplots(figsize = (8,8))
```

```python
plt.style.use('classic')
fig, axes = plt.subplots(figsize = (8,8))
```

```python
plt.style.use('_mpl-gallery')
fig, axes = plt.subplots(figsize = (8,8))
```

### Modificaciones tipo Matlab
Son modificaciones entre '', asi como se muestra a continuacion

```python
plt.style.use('dark_background')
fig, axes = plt.subplots(figsize = (8,8))
axes.plot(x+1, 'r--')
axes.plot(x+2, 'g-')
axes.plot(x+3, 'b.-')
axes.plot(x+4, 'yo:')
```

## Modificaciones tipo Pyplot

```python
plt.style.use('dark_background')
fig, axes = plt.subplots(figsize = (8,8))
# Alpha sirve para definir la transparencia de la linea
axes.plot(x+1, color= 'green', alpha=0.8)
# Linewdith para definir el ancho de la linea
axes.plot(x+2, color= 'blue', linewidth=3)
# Configuracion de colores en forma de RGB
axes.plot(x+3, color= '#66FF66')
axes.plot(x+4, color= '#CCFFCC')
```

```python
plt.style.use('dark_background')
fig, axes = plt.subplots(figsize = (8,8))
# tambien se pueden cambiar el estilo de las lineas y los marcadores
axes.plot(x+1, color= 'green', alpha=0.9, linestyle='--', marker='o' )
axes.plot(x+2, color= 'blue', linewidth=3, linestyle='dashed', marker='x')
axes.plot(x+3, color= '#66FF66', linestyle=':')
# asi mismo se puede variar el tamaño de los marcadores y el color
axes.plot(x+4, color= '#CCFFCC', marker='P', markersize=8, markerfacecolor='orange')
```
