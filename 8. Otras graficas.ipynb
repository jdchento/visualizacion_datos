## Otras graficas


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1, 50, 100)
```

```python
plt.hist(data)
plt.show()
```

```python
# Bins permite analizar la frecuencia de los datos
plt.hist(data, bins=30)
plt.show()
```

```python
# Exiten maneras de visualizar los histogramas
plt.hist(data, bins=30, histtype='step')
plt.show()
```

### Boxplot
Permite conocer donde se encuentran distribuidos mis datos

```python
plt.boxplot(data)
plt.show()
```

```python
# Para girar el grafico
plt.boxplot(data, vert=False)
plt.show()
```

```python
# Ayuda a una mejor visualizacion
plt.boxplot(data, vert=False, patch_artist=True)
plt.show()
```

```python
# Permite enfocar los datos hacia la moda
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

```python
# En caso que existan outlayers o datos no comunes
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

```python
# Para omitir datos anomalos se utiliza
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)
plt.show()
```

### Scatter
Grafica de dispersion

```python
#Rand para generar datos decimales
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2
colors = np.random.rand(N)
```

```python
print(colors)
```

```python
plt.scatter(x, y)
plt.show
```

```python
# Para añadir una tercera variable a la grafica de los puntos
plt.scatter(x, y, s=area)
plt.show
```

```python
# Para añadir una cuarta variable a la grafica de los puntos
plt.scatter(x, y, s=area, c=colors)
plt.show
```

```python
# Para cambiar la market
plt.scatter(x, y, s=area, c=colors, marker='P')
plt.show
```

```python
# Para cambiar la transparencia
plt.scatter(x, y, s=area, c=colors, marker='P', alpha=0.3)
plt.show
```
