## Heatmap
Para graficar estructuras mastriciales

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
# Analizar la correlacion de los datos
tips.corr()
```

```python
# Analizar la correlacion de los datos de forma grafica
sns.heatmap(tips.corr())
plt.show()
```

```python
# Se pueden añadir los valores
sns.heatmap(tips.corr(), annot=True)
plt.show()
```

```python
# mapas de colores
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Separacion entre las celdas
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5)
plt.show()
```

```python
# color entre las celdas
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='Black')
plt.show()
```

```python
# Asignar valores maximos y minimos
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='Black', vmin=-1, vmax=1)
plt.show()
```
