## SUBPLOT

Sirve para generar graficas en una misma salida, partiendo del concepto de filas y columnas


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
plt.subplot(1, 2, 1) # Filas, Columnas, Numero de grafico
plt.plot(x, y, 'r--') # Informacion de primer grafico
plt.subplot(1, 2, 2) # Filas, Columnas, Numero de grafico
plt.hist(y) # Informacion de primer grafico
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y, 'r--') 
plt.plot(y, x, 'b:') # Se incluye un segundo trazdo que se visualiza en el primer grafico
plt.subplot(2, 1, 2)
plt.pie(y)
plt.show()
```
