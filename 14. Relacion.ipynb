## Graficos para visualizar datos con variables con relacion

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

### Scatter

```python
# La relacion entre el total pagado y la propina que dejo el cliente

sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
# Se puede segmentar con hue

sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day')
plt.show()
```

```python
# Se puede diferenciar con style, entre cual es cena y cual almuerzo

sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
plt.show()
```

```python
# Se puede cambiar tambien el tamaño para incluir otra variable

sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.show()
```

```python
# Como la leyenda esta invadiendo el grafico se agrega la propiedad
# Es importante el orden de los siguientes metodos ya que puede llevar a que no se cree la grafica como se debe

plt.figure(figsize=(8,8)) #Aca va bien
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
# Se pueden definir los markers por defecto, comportamiento similar al de style

markers = {"Lunch":"D", "Dinner":"s"}
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

### Graficos de linea

```python
sns.lineplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
# Tiene casi los mismos parametros que scatter

sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
plt.show()
```

### relplot

Metodo que reune las graficas relacionales

```python
sns.relplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', kind='line')
plt.show()
```

```python
# Se pueden generar graficas dependientes del numero de variables
sns.relplot(data=tips, x='total_bill', y='tip', hue='time', col='day')
plt.show()
```
