## Leyendas, etiquetas, titulos, tamaños


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
# Para cambiar el tamaño de salida de las graficas se utiliza el parametro figsize
fig, axes = plt.subplots(1, 2, figsize = (5, 5))
axes[0].plot(x, y, 'b')
#Para asignar un titulo a la grafica especificada
axes[0].set_title('Relacion X - Y')
#Para asignar una etiqueta ac cada eje
axes[0].set_xlabel('Eje X')
axes[0].set_ylabel('Eje Y')

# Se puede asignar una leyenda, primero se debe definir en la grafica
# Se le ponen $ para ponerle un estilo a la letra, pero pueden ir sin ellos
axes[1].plot(y, x, 'r', label='$sin(x)$')
axes[1].set_title('Relacion Y - X')
# Para visualizar las labels se utiliza
axes[1].legend()
```

```python
# Sin el metodo orientado a objetos
plt.plot(x, y, label='sin(x)')
plt.title('Este es un titulo')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
#Para asignar la posicion de la legenda
plt.legend(loc='lower left')
```
